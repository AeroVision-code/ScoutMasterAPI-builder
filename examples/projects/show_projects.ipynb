```python
# import packages
import sys
import os
from dotenv import load_dotenv
from IPython.display import display, HTML
import json
import matplotlib.pyplot as plt
import folium

# Load environment variables from .env file
load_dotenv()

# import the API builder
sys.path.append(os.path.abspath("../../"))
from scoutmaster.api import ScoutMasterAPI
```

```python
# Scoutmaster API - Credentials
client_id = os.getenv('SM_CLIENT_ID')
client_secret = os.getenv('SM_CLIENT_SECRET')
dev= os.getenv("DEV")

# Scoutmaster API - Init
SM_API = ScoutMasterAPI(dev)
SM_API.authenticate(client_id=client_id, client_secret=client_secret)
df_projects = SM_API.projects()
df_projects
```

```python
# Create a new column for display
def display_logo_or_abbr(row):
    if row['logo_path']:
        return f'<img src="{row["logo_path"]}" width="200">'
    else:
        return f'<div style="font-weight:bold; font-size:14pt;">{row["abbreviation"]}</div>'

df_projects['display'] = df_projects.apply(display_logo_or_abbr, axis=1)

# Display table with HTML
display(HTML(df_projects[['id', 'name', 'display']].to_html(escape=False, index=False)))
```
